### Exercise 1 (Easy)
Create a simple RNN cell and pass a sequence through it.

```python
import torch
import torch.nn as nn

# Create RNN: input_size=10, hidden_size=20
# Pass a sequence of shape (seq_len=5, batch=1, input_size=10)

# We define the RNN layer
rnn = nn.RNN(input_size=10, hidden_size=20, batch_first=True)
# We create the shape (batch_size=1, seq_len=5, input_size=10)
input_seq = torch.randn(1, 5, 10)  # (batch_size, seq_len, input_size)
# We pass a sequence through the RNN
output, hidden = rnn(input_seq)
print("Output shape:", output.shape)  
print("Hidden shape:", hidden.shape)  
```

### Exercise 2 (Easy)
Create an LSTM and compare output shapes with vanilla RNN.

```python
# Create LSTM with same dimensions
# Note: LSTM returns (output, (h_n, c_n))

# We define the LSTM layer
lstm = nn.LSTM(input_size=10, hidden_size=20, batch_first=True)
# We pass the same input sequence through the LSTM
lstm_output, (h_n, c_n) = lstm(input_seq)
print("LSTM Output shape:", lstm_output.shape)
print("LSTM Hidden state shape (h_n):", h_n.shape)
print("LSTM Cell state shape (c_n):", c_n.shape)
```

### Exercise 3 (Medium)
Build an LSTM-based sentiment classifier.

```python

class LSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, num_classes):
        super().__init__()
        # Define: embedding, lstm, fc
        # We will use the last hidden state for classification
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, num_classes)

    
    def forward(self, x):
        # Embed -> LSTM -> take last hidden -> classify
        # We pass the input through the embedding layer
        embedded = self.embedding(x)
        lstm_out, (h_n, c_n) = self.lstm(embedded)
        output = self.fc(h_n[-1])
        return output

# We create a dummy input and test the model
vocab_size = 100
embed_dim = 50
hidden_dim = 20
num_classes = 2
model = LSTMClassifier(vocab_size, embed_dim, hidden_dim, num_classes)
dummy_input = torch.randint(0, vocab_size, (1, 5))  # (batch_size, seq_len)
output = model(dummy_input)
print("Model output shape:", output.shape)

```
